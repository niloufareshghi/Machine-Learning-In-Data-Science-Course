```python
import math
def isperfectsquare(m):
  a = int(math.sqrt(m))
  return a*a == m

def isfib(n):
  return isperfectsquare(5*n*n - 4) or isperfectsquare(5*n*n + 4)
```

```python
n = int(input())
name = []
for i in range(1, n+1):
  if isfib(i):
    name.append('O')
  else:
    name.append('o')
print(''.join(name))
```

```python
n, k = list(map(int, input().split()))
ls = input().split()
c = 0
for a in ls:
  if not ('4' in a or '7' in a):
    c += 1
    continue 
  count = 0
  for d in a:
    if d == '4' or d == '7':
      count += 1
  if count > k:
    continue
  else:
    c += 1
print(c)
```

```python
t = int(input())
for i in range(t):
  n = int(input())
  a = list(map(int, input().split()))
  ans = []
  x = [a[i:i + 2] for i in range(0, n, 2)] 
  for chunk in x:
    b0 = chunk[1]*-1
    b1 = chunk[0]
    ans.append(str(b0))
    ans.append(str(b1))
  print(' '.join(ans))
```
